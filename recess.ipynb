```python
import pandas as pd

# Dictionary containing days and dates
days_dates = {'Monday': '2023-07-03',
              'Tuesday': '2023-07-04',
              'Wednesday': '2023-07-05',
              'Thursday': '2023-07-06',
              'Friday': '2023-07-07'}

# Create a pandas Series
series = pd.Series(days_dates)

# Print the Series
print(series)
```

# ----------EXERCISE1-------

```python
mine_datasetds = pd.read_csv("mine.csv")
mine_datasetds
```

```python
# create Data Frame
mine_dataset = pd.DataFrame(mine_datasetds)
mine_dataset
```

```python
# Check the first five rows of the dataset
mine_dataset.head()
```

```python
# Check the last five rows of the dataset
mine_dataset.tail()
```

```python
# Get information about the dataset
mine_dataset.info()
```

```python
# Remove rows with missing values
mine_dataset.dropna(inplace=True)
```

```python
# Fill in missing values with appropriate values
mine_dataset.fillna(value=0, inplace=True)
```

```python
# Remove duplicate rows
mine_dataset.drop_duplicates(inplace=True)
```

```python
# Perform correlation analysis
correlation_matrix = mine_dataset.corr()

# Print and display the correlation matrix
print(correlation_matrix)
```

```python
# Reset the index after removing rows
mine_dataset.reset_index(drop=True, inplace=True)
mine_dataset
```
