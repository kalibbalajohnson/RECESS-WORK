## Question 2

```python
import pandas as pd

# Define the sample Series
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

addition = series1 + series2

subtraction = series1 - series2

multiplication = series1 * series2

division = series1 / series2

# Convert the Series to Python lists for better visiolization
addition_values = addition.tolist()
subtraction_values = subtraction.tolist()
multiplication_values = multiplication.tolist()
division_values = division.tolist()


# Print the results
print("Addition:")
print(addition_values)

print("\nSubtraction:")
print(subtraction_values)

print("\nMultiplication:")
print(multiplication_values)

print("\nDivision:")
print(division_values)
```
